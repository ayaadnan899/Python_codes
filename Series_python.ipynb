```python
#exercice1-Q1
serie_temp=pd.Series([22,24,19,23,25,21,20],index=["Lundi","Mardi","Mercredi","jeudi","Vendredi","Samedi","Dimanche"])
serie_temp
```

```python
#Q2 la temp de mercredi
serie_temp.loc["Mercredi"]
```

```python
#Q3  les 3 premieres
serie_temp.head(3)
```

```python
#3 dernieres
serie_temp.tail(3)
```

```python
#Q4 les temps >22
serie_temp[serie_temp>22]
```

```python
#Q5 le plus chaud,le plus froid
serie_temp.idxmax(),serie_temp.max()
```

```python
serie_temp.idxmin(),serie_temp.min()
```

```python
#Q6 tri es temps
serie_temp.sort_values()
```

```python
#nombres de jours avec temp<22
(serie_temp<22).sum()
```

```python
#tracer la courbe
plt.plot(serie_temp.index,serie_temp.values,marker="o")
plt.title("Evolution des temperatures  au cours de la semaine")
plt.xlabel("Jours de la semaine")
plt.ylabel("Temperatures (C)")
plt.grid()
plt.show()
```

```python
#exercice2
#pour creer Dictionnaire
dic={"Ville":["casablanca","marrakech","fes","agadir","rabat"],"Temp":[18.5,20.0,19.0,19.5,18.0],
    "Humd":[72,50,60,65,70],"precip":[430,250,400,300,450]}
dic
```

```python
#creer dataframe
df=pd.DataFrame(dic)
df
```

```python
#afficher colone de villes, la ligne de marrakech
df["Ville"]
```

```python
df[df["Ville"]=="marrakech"]
```

```python
#affiche les villes de temp>19
df[df["Temp"]>19]
```

```python
#villes avec le plus de preciption
df.loc[df["precip"].idxmax(),"Ville"]
```

```python
#ajouter colone capital
df["Capitale"]=df["Ville"]=="rabat"
df
```

```python
#changer les index avec noms villes
df2=df.set_index("Ville")
df2
```

```python
#tracer courbe
df["Temp"].plot(marker='o', color='b', label="Temp")
df["Humd"].plot(marker='x', color='r', label="Humd")
df["precip"].plot(marker='+', color='g', label="precip")
plt.title("Meteo")
plt.xlabel("Ville")
plt.ylabel("Temperature C")
plt.grid()
plt.show()
```
